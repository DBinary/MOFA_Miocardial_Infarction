# Traing

```python
import anndata
import itertools
import networkx as nx
import numpy as np
import pandas as pd
import scanpy as sc
import scglue
from matplotlib import rcParams
```

```python
scglue.plot.set_publication_params()
rcParams["figure.figsize"] = (4, 4)
```

```python
rna = anndata.read_h5ad("scRNA_glue.h5ad")
atac = anndata.read_h5ad("scATAC_glue.h5ad")
#
```

```python
graph = nx.read_graphml("prior.graphml.gz")
```

```python
rna
```

```python
scglue.models.configure_dataset(
    rna, "NB", use_highly_variable=True,
    use_layer="counts", use_rep="X_pca"
)
```

```python
scglue.models.configure_dataset(
    atac, "NB", use_highly_variable=True,
    use_rep="X_lsi"
)
```

```python
graph = graph.subgraph(itertools.chain(
    rna.var.query("highly_variable").index,
    atac.var.query("highly_variable").index
))
```

```python
glue = scglue.models.SCGLUEModel(
    {"rna": rna, "atac": atac}, sorted(graph.nodes),
    random_seed=0
)
```

```python
glue.compile()
```

```python
glue.fit(
    {"rna": rna, "atac": atac},
    graph, edge_weight="weight", edge_sign="sign",
    directory="glue"
)
```

```python
glue.save("glue/final.dill")
!cp /content/glue/final.dill /content/drive/MyDrive/MI
# 
```

```python
dx = scglue.models.integration_consistency(
    glue, {"rna": rna, "atac": atac}, graph,
    count_layers={"rna": "counts"}
)
dx
```

```python
import seaborn as sns
_ = sns.lineplot(x="n_meta", y="consistency", data=dx).axhline(y=0.05, c="darkred", ls="--")
```

```python
rna.obsm["X_glue"] = glue.encode_data("rna", rna)
atac.obsm["X_glue"] = glue.encode_data("atac", atac)
```

```python
rna.write("RNA_afglue.h5ad", compression="gzip")
atac.write("ATAC_afglue.h5ad", compression="gzip")
!cp /content/RNA_afgle.h5ad /content/drive/MyDrive/MI
!cp /content/ATAC_afglue.h5ad /content/drive/MyDrive/MI
```
